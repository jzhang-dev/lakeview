```python
import matplotlib.pyplot as plt
import lakeview as lv
```

```python
PACBIO_BAM_URL = "https://downloads.pacbcloud.com/public/dataset/HG002-CpG-methylation-202202/HG002.GRCh38.haplotagged.bam"
PACBIO_BAI_URL = "https://downloads.pacbcloud.com/public/dataset/HG002-CpG-methylation-202202/HG002.GRCh38.haplotagged.bam.bai"
PACBIO_BAM_PATH = "HG002_GRCh38_SNURF_haplotagged.bam"
CHROMOSOME = "chr15"
START = 24_954_010
END = 24_957_232
REGION = f"{CHROMOSOME}:{START}-{END}"

lv.helpers.download_bam(PACBIO_BAM_URL, PACBIO_BAI_URL, REGION, PACBIO_BAM_PATH)
```

```python
p = lv.SequenceAlignment.from_file(PACBIO_BAM_PATH, 'rb')
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
p.draw_alignment(ax, group_by="haplotype", show_modified_bases=True, modified_bases_kw=dict(linewidth=1.5, colormaps={("C", "m", "+"): "cividis", ("C", "m", "-"): "cividis"}))
ax.set_title("$\it{SNURF}$ differentially methylated region")
ax.set_xlim(24.953e6, 24.958e6)
ax.set_xlabel("chr15 (Mb)")
ax.xaxis.set_major_formatter(lv.util.base_formatter(unit="mb", fmt="{:.3f}"))
```
