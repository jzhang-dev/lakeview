# Interactive visualization

Lakeview provides basic support for interactive visualisation to help users navigate though the data and locate interesting regions. Interactive visualisation is designed for [Jupyter Notebooks](https://jupyter.org/) and implemented using [Jupyter Widgets](https://ipywidgets.readthedocs.io/). Interactive visualisation requires a live [IPython](https://ipython.readthedocs.io/en/stable/index.html) kernel. If you are viewing the static HTML version of this page, please use the button below to run this tutorial in [Binder](https://mybinder.org/):

[![Binder](https://mybinder.org/badge_logo.svg)](https://mybinder.org/v2/gh/jzhang-dev/lakeview/HEAD?labpath=docs%2Ftutorials%2Fgenome_viewer.ipynb)

```python
import lakeview as lv

BAM_PATH = "../../tests/data/SKBR3_Illumina_550bp_pcrFREE.bam"
REGION = "17:64,040,802-64,045,633"

# Load data from the BAM file
painter = lv.SequenceAlignment.from_file(BAM_PATH, region=REGION)
# Create a GenomeViewer with two tracks
gv = lv.GenomeViewer(tracks=2, height_ratios=(1, 6), figsize=(12, 10))

# Draw alignment pileup in the first track
painter.draw_pileup(gv.axes[0])

# Draw aligned segments in the second track
painter.draw_alignment(gv.axes[1])
```
